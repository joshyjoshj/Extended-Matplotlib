```python
data = {'Year': [2000, 2001, 2002, 2003, 2004, 2005],
        'Alpha': [5, 2, 4, 1, 5, 3],
        'Bravo': [4, 1, 5, 2, 4, 5],
        'Charlie': [3, 5, 1, 4, 2, 4],
        'Delta': [2, 4, 3, 5, 1, 2],
        'Echo': [1, 3, 2, 3, 3, 1]}

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

colors = ['#233d4d', '#fe7f2d', '#FCCA46', '#81AE86', '#619B8A']

fig, ax = plt.subplots(figsize=(10,7))

for column, color in zip(df.columns, colors):
    y = df[column].values
    x = df.index.values

    # Plot points
    ax.plot(x, y, 'o', color=color, label=column)

    for i in range(len(x)-1):
        #  mid
        midpoint = (x[i] + x[i+1]) / 2

        #sigmoid function 
        sigmoid_x = np.linspace(x[i], x[i+1], 500)
        sigmoid_y = y[i] + (y[i+1] - y[i]) * expit(13 * (sigmoid_x - midpoint))

        ax.plot(sigmoid_x, sigmoid_y, linestyle='-', color=color, linewidth=4)

ax.invert_yaxis()

ax.legend(loc='best')
ax.grid()

plt.show()
```
