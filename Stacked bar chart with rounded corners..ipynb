```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch, Circle
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35  
```

```python
data = {
    'Country': ['Norway', 'Denmark', 'Sweden'],
    'Value': [5, 4, 13]
}
df = pd.DataFrame(data)

# Plot
plt.subplots(figsize=(10, 4))
ax = plt.barh(df.Country, df.Value, color='skyblue')

new_patches = []
for patch in reversed(ax.patches):
    bb = patch.get_bbox()
    color = patch.get_facecolor()
    p_bbox = patches.FancyBboxPatch((bb.xmin, bb.ymin),
                            abs(bb.width), abs(bb.height),
                            boxstyle="round,pad=-0.004,rounding_size=1",
                            ec="none", fc=color,
                            mutation_aspect=0.49
                            )
    patch.remove()
    new_patches.append(p_bbox)

for patch in new_patches:
    plt.gca().add_patch(patch)

plt.grid(False)
plt.tick_params(axis=u'both', which=u'both', length=0)
plt.tight_layout()
plt.show()
```

```python
data = {
    'Country': ['Norway', 'Denmark', 'Sweden'],
    'Value': [5, 4, 13]
}
df = pd.DataFrame(data)


fig, ax = plt.subplots(figsize=(5, 2))

#colours
colors = {'Norway': '#f5f3bb', 'Denmark': '#86ba90', 'Sweden': '#df2935'}

bars = ax.barh(df.Country, df.Value, color=[colors[country] for country in df.Country])

new_patches = []
for bar in reversed(bars):
    bb = bar.get_bbox()
    color = bar.get_facecolor()
    p_bbox = FancyBboxPatch((bb.xmin, bb.ymin),
                            abs(bb.width), abs(bb.height),
                            boxstyle="round,pad=-0.0040,rounding_size=1",
                            ec="none", fc=color,
                            mutation_aspect=0.49
                            )
    bar.remove()
    new_patches.append(p_bbox)

for patch in new_patches:
    ax.add_patch(patch)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tick_params(axis=u'both', which=u'both', length=0)
plt.tight_layout()
plt.show()
```

```python
data = {
    'Country': ['Denmark', 'Norway', 'Sweden'],
    'Value': [4, 5, 13]
}
df = pd.DataFrame(data)

data2 = {
    'Country': ['Denmark', 'Norway', 'Sweden'],
    'Value': [10, 8, 15]
}
df2 = pd.DataFrame(data2)


colors = {'Norway': '#809bce', 'Denmark': '#86ba90', 'Sweden': '#df2935'}

fig, ax = plt.subplots(figsize=(18, 6))

bars2 = ax.barh(df2.Country, df2.Value, color=[colors[country] for country in df2.Country], alpha=0.5)
new_patches2 = []
for bar, value in zip(reversed(bars2), reversed(df2.Value)):
    bb = bar.get_bbox()
    color = bar.get_facecolor()
    p_bbox = FancyBboxPatch((bb.xmin, bb.ymin),
                            abs(bb.width), abs(bb.height),
                            boxstyle="round,pad=-0.0040,rounding_size=1",
                            ec="none", fc=color,
                            mutation_aspect=0.49
                            )
    bar.remove()
    new_patches2.append(p_bbox)
    ax.text(bb.xmax - 0.5, bb.ymin + bb.height / 2, str(value), va='center', color='white', fontsize = 50, ha = "right",
           bbox = dict(boxstyle=f"circle,pad=0.1", fc="none", ec = "white", lw = 2))

for patch in new_patches2:
    ax.add_patch(patch)

bars = ax.barh(df.Country, df.Value, color=[colors[country] for country in df.Country])
new_patches = []
for bar, value in zip(reversed(bars), reversed(df.Value)):
    bb = bar.get_bbox()
    color = bar.get_facecolor()
    p_bbox = FancyBboxPatch((bb.xmin, bb.ymin),
                            abs(bb.width), abs(bb.height),
                            boxstyle="round,pad=-0.0040,rounding_size=1",
                            ec="none", fc=color,
                            mutation_aspect=0.49
                            )
    bar.remove()
    new_patches.append(p_bbox)
    ax.text(bb.xmax - 0.5, bb.ymin + bb.height / 2, str(value), va='center', color='white', fontsize = 50, ha = "right",
           bbox = dict(boxstyle=f"circle,pad=0.1", fc="none", ec = "white", lw = 2))

for patch in new_patches:
    ax.add_patch(patch)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_visible(False)
plt.tick_params(axis=u'both', which=u'both', length=0)
ax.tick_params(axis='y', labelsize=30)

ax.annotate('2004', xy=(12.05, 2.3), xytext=(12.05, 2.8),
            arrowprops=dict(facecolor='black', arrowstyle = "]-"), ha='center',
            size = 25
           )

ax.annotate('2022', xy=(14.05, 2.3), xytext=(14.05, 2.8),
            arrowprops=dict(facecolor='black', arrowstyle = "]-"), ha='center',
            size = 25
           )

plt.tight_layout()
plt.show()
```
